```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('/outlet_sales.xlsx')
```

```python
# Checking for duplicates
data.duplicated().any()
```

```python
data = data.dropna()
```

```python
data
```

```python
#Analysis
#Total Unit Ordered
Total_unit_ordered = data['Order_Quantity'].sum()
```

```python
Total_unit_ordered
```

```python
#Total Revenue
Total_Revenue = data['Revenue'].sum()
```

```python
Total_Revenue
```

```python
#Total Profits
Total_Profit = data['Profit'].sum()
```

```python
Total_Profit
```

```python
#Product category over revenue generated
data.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
#BarH showing charts
data.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False).plot(color='r', kind='barh')
```

```python
#using pivot table, display the country, and the total profit by gender
data.pivot_table(index='Country', columns='Customer_Gender', values='Profit', aggfunc='sum')
```

```python
#Using pie Chart display the age group with their order quantity
data.groupby('Age_Group')['Order_Quantity'].sum().sort_values(ascending=False).plot(kind='pie')
```

```python
#using the scatterplot, display the relationship between revenue and the profits.
data.plot(x='Revenue', y='Profit',  kind='scatter')
```
