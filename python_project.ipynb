This is a descriptive data analysis project on the outlet sales. Operations started ein 2011 and below are the metrics that display their performances across 6 years of operations.

```python
#Libraries for Data Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Tasks**

2. Analyse the data

b.) The total Revenue
c.) Total Profit

3. Product Category with the most revenue - use BarH
4. using pivot table, display the country, and the total profit by gender
5. Using pie Chart display the age group with their order quantity
6. using the scatterplot, display the relationship between revenue and the profits.

```python
data= pd.read_excel('/content/outlet_sales.xlsx')
```

```python
data
```

```python
# Checking for duplicates
data.duplicated().any()
```

```python
data = data.dropna()
```

```python
data
```

```python
#Analysis
#Total Unit Ordered
Total_unit_ordered = data['Order_Quantity'].sum()
```

```python
Total_unit_ordered
```

```python
#Total Revenue
Total_Revenue = data['Revenue'].sum()
```

```python
Total_Revenue
```

```python
#Total Profits
Total_Profit = data['Profit'].sum()
```

```python
Total_Profit
```

```python
#Product category over revenue generated
data.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
#BarH showing charts
data.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False).plot(color='r', kind='barh')
plt.title('Product Category with the most revenue')
plt.xlabel('Revenue')
plt.ylabel('Product Category')
plt.grid()
plt.show()
```

```python
#using pivot table, display the country, and the total profit by gender
data.pivot_table(index='Country', columns='Customer_Gender', values='Profit', aggfunc='sum')
```

```python
#Using pie Chart display the age group with their order quantity
data.groupby('Age_Group')['Order_Quantity'].sum().sort_values(ascending=False).plot(kind='pie', autopct = '%1.0f%%')
plt.title('Age Group with their Order Quantity')
plt.show()
```

```python
#using the scatterplot, display the relationship between revenue and the profits.
data.plot(x='Revenue', y='Profit',  kind='scatter')
plt.title('Revenue vs Profit')
plt.xlabel('Revenue')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
#Top 5 states with their revenue generated
data.groupby('State')['Revenue'].sum().sort_values(ascending=True).head(5)
```

```python
#using bar chart to show the top 5 states with their revenue generated .
data.groupby('State')['Revenue'].sum().sort_values(ascending=True).head(5).plot(kind='barh')
plt.title('Top 5 states with their revenue generated')
plt.xlabel('Revenue')
plt.ylabel('State')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
top_5_states = data.groupby('State')['Revenue'].sum().sort_values(ascending=False).head(5)

# Create horizontal bar chart
plt.figure(figsize=(8, 5))
bars = plt.barh(top_5_states.index, top_5_states.values, color='skyblue')

# Add labels inside the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width - (width * 0.1), bar.get_y() + bar.get_height()/2, f'${width:,.0f}',
             ha='right', va='center', color='white', fontweight='bold')

# Add titles and labels
plt.title('Top 5 States by Revenue Generated', fontsize=14)
plt.xlabel('Revenue', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.gca().invert_yaxis()  # Ensures the state with the highest revenue is at the top

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
top_5_states = data.groupby('State')['Revenue'].sum().sort_values(ascending=False).head(5)

# Create horizontal bar chart
plt.figure(figsize=(8, 5))
bars = plt.barh(top_5_states.index, top_5_states.values, color='skyblue')

# Add labels using plt.text() without a loop
plt.text(top_5_states.values.sum(), str(len(top_5_states)),
         [f'{val:,.0f}' for val in top_5_states.values],
         ha='right', va='center', color='white', fontweight='bold')

# Add titles and labels
plt.title('Top 5 States by Revenue Generated', fontsize=14)
plt.xlabel('Revenue', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.gca().invert_yaxis()  # Ensures the highest revenue state is at the top

# Show plot
plt.show()
```
